```python
import cv2
import glob
import tensorflow as tf
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm
```

## Constants

```python
IMG_SIZE = 128
SOURCE = "train_images/"
RES = f"train_images_res_{IMG_SIZE}/"
```

## Get IMG

```python
train_csv = "train.csv"
train_df = pd.read_csv(train_csv)
images = train_df['image']
full_img = list(map(RES.__add__, images))

dataset = tf.data.Dataset.from_tensor_slices(full_img)
```

```python
def show(image):
    plt.imshow(image)
    plt.show()
    
    
show_amt = 3
images = []
for i,img in enumerate(dataset.shuffle(show_amt).take(show_amt)):
    im = cv2.imread(img.numpy().decode())
    image = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    images.append(image)
    show(image)
    
```

## AVG IMG

```python
image_data = []

for img in tqdm(dataset):
    this_image = cv2.imread(img.numpy().decode())
    image_data.append(this_image)

avg_image = image_data[0]
for i in tqdm(range(len(image_data))):
    if i == 0:
        pass
    else:
        alpha = 1.0/(i + 1)
        beta = 1.0 - alpha
        avg_image = cv2.addWeighted(image_data[i], alpha, avg_image, beta, 0.0)

cv2.imwrite('avg.png', avg_image)
avg_image = cv2.imread('avg.png')
plt.imshow(avg_image)
plt.show()
```

## AVG IMG per Species

```python
species_labels = list(train_df.species.unique())
species = []
for sp in species_labels:
    species.append(train_df.loc[train_df['species'] == sp]['image'])
    
def make_avg(species,arr):
    full_img = list(map(RES.__add__, arr))
    
    image_data = []
    for img in full_img:
        this_image = cv2.imread(img)
        image_data.append(this_image)

    avg_image = image_data[0]
    for i in tqdm(range(len(image_data))):
        if i == 0:
            pass
        else:
            alpha = 1.0/(i + 1)
            beta = 1.0 - alpha
            avg_image = cv2.addWeighted(image_data[i], alpha, avg_image, beta, 0.0)
    
    print(f'avg of {species}')
    cv2.imwrite(f'avg_imgs/avg_{species}.jpg', avg_image)
    avg_image = cv2.imread(f'avg_imgs/avg_{species}.jpg')
    plt.imshow(avg_image)
    plt.show()

for i,sp in enumerate(species_labels):
    make_avg(sp, species[i])
```
